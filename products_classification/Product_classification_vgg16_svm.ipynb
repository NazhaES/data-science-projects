```python
import os
import glob

# Path to the "train" directory
train_dir = "train2"

# Get a list of all subdirectories in the "train" directory
subdirectories = [subdir for subdir in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, subdir))]

# Iterate through each subdirectory and count the number of images
total_images = 0
for subdir in subdirectories:
    # Count the number of images in the current subdirectory
    images_in_subdir = len(glob.glob(os.path.join(train_dir, subdir, "*")))

    # Print the number of images in the current subdirectory
    print(f"Number of images in {subdir}: {images_in_subdir}")

    # Add the number of images in the current subdirectory to the total count
    total_images += images_in_subdir

# Print the total number of images in the "train" directory
print(f"Total number of images in 'train' directory: {total_images}")
```

```python
# Load pre-trained VGG16 model
vgg_model = VGG16(weights='imagenet', include_top=False)

# Function to extract features using VGG16
def extract_features(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_data = image.img_to_array(img)
    img_data = np.expand_dims(img_data, axis=0)
    img_data = preprocess_input(img_data)
    vgg_features = vgg_model.predict(img_data)
    vgg_features_flattened = vgg_features.flatten()
    return vgg_features_flattened

# Directory containing subdirectories, each named after a class
data_dir = 'train2'

# Augment data
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

# Get the class names from subdirectory names
class_names = os.listdir(data_dir)

# Create a dictionary mapping class names to numeric labels
class_to_label = {class_name: i for i, class_name in enumerate(class_names)}

# Initialize lists to store features and labels
features = []
labels = []

# Iterate over each class subdirectory
for class_name in class_names:
    class_dir = os.path.join(data_dir, class_name)
    # Iterate over each image in the class subdirectory
    for img_name in os.listdir(class_dir):
        img_path = os.path.join(class_dir, img_name)
        # Extract features and append to the features list
        img_features = extract_features(img_path)
        features.append(img_features)
        # Append the corresponding label to the labels list
        labels.append(class_to_label[class_name])

        # Augment data and add to features/labels
        img = image.load_img(img_path, target_size=(224, 224))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        for batch in datagen.flow(img_array, batch_size=1):
            features.append(vgg_model.predict(preprocess_input(batch))[0].flatten())
            labels.append(class_to_label[class_name])
            break
# Convert lists to arrays
features_array = np.array(features)
labels_array = np.array(labels)

# Save features and labels to files
np.save('features100.npy', features_array)
np.save('labels100.npy', labels_array)
```

```python
# Load features and labels from files
features_array = np.load('features100.npy')
labels_array = np.load('labels100.npy')
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_array, labels_array, test_size=0.2, random_state=42)

print("Total training examples: ", len(X_train))
print("Total test examples: ", len(X_test))
```

```python
X_train=X_train/255 ## Normalizing train data
X_test=X_test/255 ## Normalizing test data
```

```python
# Train SVM classifier
svm_classifier = SVC(kernel='linear', C=0.1)
```

```python
svm_classifier.fit(X_train, y_train)
```

```python
import joblib

# Assuming svm_classifier is your trained SVC model
joblib.dump(svm_classifier, 'svm_classifier100_2.pkl')
```

```python
svm_classifier = joblib.load('svm_classifier100_2.pkl')
```

```python
# Predict labels for test set
y_pred = svm_classifier.predict(X_test)
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
